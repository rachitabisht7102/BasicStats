```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
cars_df= pd.read_csv("wc-at.csv")
```

```python
cars_df
```

```python
plt.figure(figsize=(10,5))
plt.title("Distribution")
sn.distplot(cars_df.Waist,label="cars_df.csv")
plt.xlabel("Waist")
plt.ylabel("Density")
plt.plot()
```

```python
plt.figure(figsize=(10,5))
plt.title("Distribution")
sn.distplot(cars_df.AT,label="cars_df.csv")
plt.xlabel("Waist")
plt.ylabel("Density")
plt.plot()
```
